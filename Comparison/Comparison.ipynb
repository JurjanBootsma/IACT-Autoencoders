# Comparison of different cleaning algorithms


This notebook explores the different distributions of the IoU and normalized MSE.

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Load IoU distributions**

```python
IOU_tc = np.load("../Tailcuts/data_output/IOU_tc_dis.npy")
IOU_seg = np.load("../Segmentation/data_output/IOU_seg_dis.npy")
IOU_reg = np.load("../Regression/data_output/IOU_reg_dis.npy")
IOU_hyb = np.load("../Hybrid/data_output/IOU_hybrid_dis.npy")
```

```python
linewidth = 1

plt.figure()
#plt.hist(IOU_tc, bins=20, range=(0,1), density=True, histtype='step', linewidth=linewidth, label='Tailcuts')
plt.hist(IOU_tc, bins=20, range=(0,1), density=True, alpha=0.2, linewidth=linewidth, label='Tailcuts')
plt.hist(IOU_seg, bins=20, range=(0,1), density=True, histtype='step', linewidth=linewidth, label='Segmentation')
plt.hist(IOU_reg, bins=20, range=(0,1), density=True, histtype='step', linewidth=linewidth, label='Regression')
plt.hist(IOU_hyb, bins=20, range=(0,1), density=True, histtype='step', linewidth=linewidth, label='Hybrid')
plt.xlabel("IoU")
plt.ylabel("Probability density")
plt.legend()
plt.show()
```

**Load MSE distributions**

```python
MSE_tc = np.load("../Tailcuts/data_output/MSE_tc_dis.npy")
MSE_seg = np.load("../Segmentation/data_output/MSE_seg_dis.npy")
MSE_reg = np.load("../Regression/data_output/MSE_reg_dis.npy")
MSE_hyb = np.load("../Hybrid/data_output/MSE_hybrid_dis.npy")
```

```python
linewidth = 1

plt.figure()
#plt.hist(MSE_tc, bins=20, range=(0,0.001), density=True, histtype='step', linewidth=linewidth, label='Tailcuts')
plt.hist(MSE_tc, bins=20, range=(0,0.001), density=True, alpha=0.2, linewidth=linewidth, label='Tailcuts')
plt.hist(MSE_seg, bins=20, range=(0,0.001), density=True, histtype='step', linewidth=linewidth, label='Segmentation')
plt.hist(MSE_reg, bins=20, range=(0,0.001), density=True, histtype='step', linewidth=linewidth, label='Regression')
plt.hist(MSE_hyb, bins=20, range=(0,0.001), density=True, histtype='step', linewidth=linewidth, label='Hybrid')
plt.xlabel("MSE")
plt.ylabel("Probability density")
plt.legend()
plt.show()
```

**Plot together**

```python
# For report

fig, axs = plt.subplots(1,2,figsize=(15,5))

axs[0].hist(IOU_tc, bins=20, range=(0,1), density=True, alpha=0.2, linewidth=linewidth, label='Tailcuts')
axs[0].hist(IOU_seg, bins=20, range=(0,1), density=True, histtype='step', linewidth=linewidth, label='Segmentation')
axs[0].hist(IOU_reg, bins=20, range=(0,1), density=True, histtype='step', linewidth=linewidth, label='Regression')
axs[0].hist(IOU_hyb, bins=20, range=(0,1), density=True, histtype='step', linewidth=linewidth, label='Hybrid')
axs[0].set_xlabel("IoU")
axs[0].set_ylabel("Probability density")
axs[0].legend()
axs[0].set_title("IoU distributions")

axs[1].hist(MSE_tc, bins=20, range=(0,0.001), density=True, alpha=0.2, linewidth=linewidth, label='Tailcuts')
axs[1].hist(MSE_seg, bins=20, range=(0,0.001), density=True, histtype='step', linewidth=linewidth, label='Segmentation')
axs[1].hist(MSE_reg, bins=20, range=(0,0.001), density=True, histtype='step', linewidth=linewidth, label='Regression')
axs[1].hist(MSE_hyb, bins=20, range=(0,0.001), density=True, histtype='step', linewidth=linewidth, label='Hybrid')
axs[1].set_xlabel("MSE")
axs[1].set_ylabel("Probability density")
axs[1].set_title("MSE distributions")
axs[1].legend()

plt.show()
fig.savefig("images/Distributions.pdf")
```

**Plots for presentation**
